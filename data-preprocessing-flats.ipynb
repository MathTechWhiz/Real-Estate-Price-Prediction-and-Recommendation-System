```python
# we have to normalize the price column bc it have both lac and cr so all convert into into cr 
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
df.head()
```

```python
#we have to treat the price column 
def treat_price(x):
    if type(x)== float:
        return x
    else:
        if x[1]== 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price']=df['price'].str.split(' ').apply(treat_price)
#df['price'].describe()
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
